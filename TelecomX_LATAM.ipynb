#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
```

```python
datos = pd.read_json('/content/TelecomX_Data.json')
datos.head()
```

```python
print(datos.info)
```

```python
datos.dtypes
```

#ðŸ”§ TransformaciÃ³n

```python
datos = pd.concat([
    datos.drop(['customer', 'phone', 'internet', 'account'], axis=1),
    datos['customer'].apply(pd.Series),
    datos['phone'].apply(pd.Series),
    datos['internet'].apply(pd.Series),
    datos['account'].apply(pd.Series)
], axis=1)
datos
```

```python
charges_expanded = datos['Charges'].apply(pd.Series)
datos = pd.concat([datos.drop('Charges', axis=1), charges_expanded], axis=1)
datos
```

```python
print(datos.isnull().sum())
```

```python
print(f"Filas duplicadas: {datos.duplicated().sum()}")
```

```python
print(datos.dtypes)
```

```python
datos['Total'] = pd.to_numeric(datos['Total'], errors='coerce')
datos.dtypes
```

```python
print(datos.isnull().sum())
```

```python
print(datos[datos['Total'].isnull()])
```

```python
datos['Total'].fillna(0, inplace=True)
```

```python
print(datos['Total'].isnull().sum())
```

```python
datos.columns = datos.columns.str.lower().str.replace(' ', '_')
datos
```

```python
datos.dtypes
```

#ðŸ“Š Carga y anÃ¡lisis

```python
datos['cuentas_diarias'] = datos['monthly'] / 30
datos.head()
```

```python
print(datos[['monthly', 'cuentas_diarias']].head())
```

```python
datos.dtypes
```

```python
# Definir diccionario para mapear texto a nÃºmeros
map_si_no = {
    'yes': 1,
    'no': 0,
    'sÃ­': 1,
    'no': 0
}

# Lista con nombre cambiado a columnas_binarias
columnas_binarias = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']

for col in columnas_binarias:
    datos[col] = datos[col].str.lower().map(map_si_no)
```

```python
datos.head()
```

```python
datos.rename(columns={
    'customerid': 'id_cliente',
    'gender': 'genero',
    'seniorcitizen': 'adulto_mayor',
    'partner': 'tiene_pareja',
    'dependents': 'tiene_dependientes',
    'tenure': 'meses_contrato',
    'phoneservice': 'servicio_telefono',
    'multiplelines': 'lineas_adicionales',
    'internetservice': 'servicio_internet',
    'onlinesecurity': 'seguridad_online',
    'onlinebackup': 'respaldo_online',
    'deviceprotection': 'proteccion_dispositivos',
    'techsupport': 'soporte_tecnico',
    'streamingtv': 'streaming_tv',
    'streamingmovies': 'streaming_peliculas',
    'contract': 'tipo_contrato',
    'paperlessbilling': 'facturacion_sin_papel',
    'paymentmethod': 'metodo_pago',
    'monthly': 'facturacion_mensual',
    'total': 'facturacion_total',
    'cuentas_diarias': 'facturacion_diaria',
    'churn': 'tasa_abandono'
}, inplace=True)
```

```python
datos.head()
```

#ðŸ“Š Carga y anÃ¡lisis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definimos las columnas numÃ©ricas que vamos a analizar
numericas = [
    'tasa_abandono', 'adulto_mayor', 'tiene_pareja', 'tiene_dependientes', 'meses_contrato',
    'servicio_telefono', 'facturacion_sin_papel', 'facturacion_mensual', 'facturacion_total', 'facturacion_diaria'
]
```

```python
estadisticas = datos[numericas].describe()
medianas = datos[numericas].median()
desviaciones = datos[numericas].std()
```

```python
print("=== EstadÃ­sticas bÃ¡sicas ===")
print(estadisticas)
print("\n=== Medianas ===")
print(medianas)
print("\n=== Desviaciones estÃ¡ndar ===")
print(desviaciones)
```

##AnÃ¡lisis Descriptivo

```python
sns.set(style="whitegrid")

# Histogramas para variables numÃ©ricas continuas
plt.figure(figsize=(15,10))
for i, col in enumerate(['tasa_abandono', 'meses_contrato', 'facturacion_mensual', 'facturacion_total', 'facturacion_diaria']):
    plt.subplot(2,3,i+1)
    sns.histplot(datos[col], bins=30, kde=True, color='skyblue')
    plt.title(f'DistribuciÃ³n de {col}')
plt.tight_layout()
plt.show()
```

```python
#Columnas binarias
binarias = ['adulto_mayor', 'tiene_pareja', 'tiene_dependientes', 'servicio_telefono', 'facturacion_sin_papel']

plt.figure(figsize=(15,7))
for i, col in enumerate(binarias):
    plt.subplot(2,3,i+1)
    sns.countplot(x=col, data=datos, palette='Set2')
    plt.title(f'Conteo de {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='tasa_abandono', y='facturacion_mensual', data=datos)
plt.title('FacturaciÃ³n mensual segÃºn tasa de abandono')
plt.show()
```

##DistribuciÃ³n de evasiÃ³n

```python
conteo = datos['tasa_abandono'].value_counts().sort_index()

print("DistribuciÃ³n de tasa_abandono:")
print(conteo)

plt.figure(figsize=(6,6))
colors = ['#2ecc71', '#e74c3c']
plt.pie(conteo.values, labels=['Permanecieron', 'Abandonaron'], autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Porcentaje de clientes segÃºn tasa_abandono')
plt.axis('equal')
plt.show()
```

##Recuento de evasiÃ³n por variables categÃ³ricas

```python
# Lista de variables categÃ³ricas relevantes para explorar
categoricas = ['genero', 'tipo_contrato', 'metodo_pago', 'lineas_adicionales', 'servicio_internet']

# ConfiguraciÃ³n general de grÃ¡ficos
sns.set(style="whitegrid")
plt.figure(figsize=(18, 12))

for i, col in enumerate(categoricas, 1):
    plt.subplot(2, 3, i)
    # Calculamos la tasa promedio de abandono por categorÃ­a
    tasa_por_categoria = datos.groupby(col)['tasa_abandono'].mean().sort_values(ascending=False)

    # Barplot de tasa promedio de abandono
    sns.barplot(x=tasa_por_categoria.values, y=tasa_por_categoria.index, palette='coolwarm')
    plt.title(f'Tasa promedio de abandono por {col}')
    plt.xlabel('Tasa promedio de abandono')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

##Conteo de evasiÃ³n por variables numÃ©ricas

```python
numericas = ['facturacion_total', 'meses_contrato', 'facturacion_mensual', 'facturacion_diaria']

sns.set(style="whitegrid")
plt.figure(figsize=(16, 10))

for i, col in enumerate(numericas, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='tasa_abandono', y=col, data=datos, palette=['#2ecc71', '#e74c3c'])
    plt.title(f'DistribuciÃ³n de {col} segÃºn evasiÃ³n')
    plt.xlabel('Tasa de abandono (0=No, 1=SÃ­)')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```
