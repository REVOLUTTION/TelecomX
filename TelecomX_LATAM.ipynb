#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
```

```python
datos = pd.read_json('/content/TelecomX_Data.json')
datos.head()
```

```python
print(datos.info)
```

```python
datos.dtypes
```

#ðŸ”§ TransformaciÃ³n

```python
datos = pd.concat([
    datos.drop(['customer', 'phone', 'internet', 'account'], axis=1),
    datos['customer'].apply(pd.Series),
    datos['phone'].apply(pd.Series),
    datos['internet'].apply(pd.Series),
    datos['account'].apply(pd.Series)
], axis=1)
datos
```

```python
charges_expanded = datos['Charges'].apply(pd.Series)
datos = pd.concat([datos.drop('Charges', axis=1), charges_expanded], axis=1)
datos
```

```python
print(datos.isnull().sum())
```

```python
print(f"Filas duplicadas: {datos.duplicated().sum()}")
```

```python
print(datos.dtypes)
```

```python
datos['Total'] = pd.to_numeric(datos['Total'], errors='coerce')
datos.dtypes
```

```python
print(datos.isnull().sum())
```

```python
print(datos[datos['Total'].isnull()])
```

```python
datos['Total'].fillna(0, inplace=True)
```

```python
print(datos['Total'].isnull().sum())
```

```python
datos.columns = datos.columns.str.lower().str.replace(' ', '_')
datos
```

```python
datos.dtypes
```

#ðŸ“Š Carga y anÃ¡lisis

```python
datos['cuentas_diarias'] = datos['monthly'] / 30
datos.head()
```

```python
print(datos[['monthly', 'cuentas_diarias']].head())
```

```python
datos.dtypes
```

```python
# Definir diccionario para mapear texto a nÃºmeros
map_si_no = {
    'yes': 1,
    'no': 0,
    'sÃ­': 1,
    'no': 0
}

# Lista con nombre cambiado a columnas_binarias
columnas_binarias = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']

for col in columnas_binarias:
    datos[col] = datos[col].str.lower().map(map_si_no)
```

```python
datos.head()
```

```python
datos.rename(columns={
    'customerid': 'id_cliente',
    'gender': 'genero',
    'seniorcitizen': 'adulto_mayor',
    'partner': 'tiene_pareja',
    'dependents': 'tiene_dependientes',
    'tenure': 'meses_contrato',
    'phoneservice': 'servicio_telefono',
    'multiplelines': 'lineas_adicionales',
    'internetservice': 'servicio_internet',
    'onlinesecurity': 'seguridad_online',
    'onlinebackup': 'respaldo_online',
    'deviceprotection': 'proteccion_dispositivos',
    'techsupport': 'soporte_tecnico',
    'streamingtv': 'streaming_tv',
    'streamingmovies': 'streaming_peliculas',
    'contract': 'tipo_contrato',
    'paperlessbilling': 'facturacion_sin_papel',
    'paymentmethod': 'metodo_pago',
    'monthly': 'facturacion_mensual',
    'total': 'facturacion_total',
    'cuentas_diarias': 'facturacion_diaria',
    'churn': 'tasa_abandono'
}, inplace=True)
```
